# Importing the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Importing the dataset

```python
dataset = pd.read_csv("Drug_Effectiveness.csv")
X = dataset.iloc[ : , 0:-1].values
y = dataset.iloc[ : , -1].values
```

# Simple Linear Regression

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X,y)
```

##Predicting a new result

```python
print(regressor.predict([[12]]))
```

```python
y_pred = regressor.predict(X)
```

```python
print(y_pred)
```

## Visualizing the Results

```python
plt.scatter(X, y, color = "red")
# plt.scatter(X, y_pred, color = "blue")
plt.plot(X, y_pred, color = "blue")
plt.title("Dosage VS Effectiveness")
plt.xlabel("Dosage (mg)")
plt.ylabel("Effectiveness (%)")
plt.show()
```

##Decision Tree Regression

##Training the decision tree regression model

```python
from sklearn.tree import DecisionTreeRegressor
decisionRegressor= DecisionTreeRegressor()
decisionRegressor.fit(X,y)
```

##Predicting a new result


```python
print(decisionRegressor.predict([[12]]))
```

##Visualizing the decision tree regression results

```python
plt.scatter(X,y, color="red")
plt.plot(X,decisionRegressor.predict(X),color="blue")
plt.title("Dosage VS Effectiveness")
plt.xlabel("Dosage (mg)")
plt.ylabel("Effectiveness (%)")
plt.show()


```
